# **Model Implementation :** *Part-2(Outliers Capped)*

We saw the model performances with out treating any of the outliers. Now, let's check the performance with outliers capping.

**Approach :** We follow the Winsorization approach for this. We will calculate the 99th & 1st percentile and make them the upper and lower cut-off. If any observation have values outside of these bounds, then they will be replaced by the cut-off values.

```python
# checking the 1st and 99th percentile
X[numeric_features].describe(percentiles=[0.01,0.05,0.95,0.99])
```

```python
# capping outlier values
for i in numeric_features:   # iterating through numeric columns

  upper_limit = X[i].quantile(0.99)   # 99th percentile
  lower_limit = X[i].quantile(0.01)   # 1th percentile

  X[i] = np.where(X[i] > upper_limit,
                  upper_limit,    # capping values with upper limit
                  np.where(X[i] < lower_limit,
                           lower_limit,   # capping with lower limit
                           X[i]))
```

```python
# after outlier capping
X[numeric_features].describe(percentiles=[0.01,0.05,0.95,0.99])
```

```python
# doing the train_test_split on our data
X_train, X_test, y_train, y_test = train_test_split( X, y , test_size = 0.3, random_state = 7)


# scaling
scaler = MinMaxScaler()

# transform data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### <b>Oversampling using smote:

```python
# oversampling X_train and y_train with SMOTE
X_train_smote , y_train_smote = smote.fit_resample(X_train_scaled, y_train)
```

We will be training different classifiers and then check their performances against unseen test data.

```python
from pandas.core.common import random_state
# creating instances for different models
log_reg = LogisticRegression(fit_intercept=True, random_state = 7)
knn_model = KNeighborsClassifier(n_neighbors = 5)
sv_clf = SVC(kernel='sigmoid')
dt_clf = DecisionTreeClassifier( random_state=7)
gnb_clf = GaussianNB()
mnb_clf = MultinomialNB()
rf_clf = RandomForestClassifier(random_state=7)
sgb_model = SGDClassifier(random_state= 7)
gb_model = GradientBoostingClassifier(random_state= 7)
adb_model = AdaBoostClassifier(random_state= 7)
lgbm_clf = LGBMClassifier(random_state= 7)
xgb_clf = XGBClassifier(random_state= 7)
```

```python
classifiers = [
          [log_reg,'Logistic Regression model'],
          [knn_model, 'K-Nearest Neighbours'],
          [sv_clf,'SVC'],
          [dt_clf,'Decision Tree Classifier'],
          [gnb_clf,'GaussianNB'],
          [mnb_clf, 'MultinomialNB'],
          [rf_clf,'Random Forest'],
          [sgb_model,'SGD Classifier'],
          [gb_model, 'Gradient Boost Classifier'],
          [adb_model, 'Adaboost Classifier'],
          [lgbm_clf,'LGBM Classifier'],
          [xgb_clf,'XGB Classifier']
        ]
```

```python
Accuracy = []
Recall = []
Precision = []
F1_score = []
ROC_AUC = []

for clfs in classifiers:
    model = clfs[0]
    model.fit(X_train_smote, y_train_smote)

    y_pred = model.predict(X_test_scaled)

    Accuracy.append(round(accuracy_score(y_test, y_pred) * 100, 2))
    Recall.append(round(recall_score(y_test, y_pred) * 100, 2))
    Precision.append(round(precision_score(y_test, y_pred) * 100, 2))
    F1_score.append(round(f1_score(y_test, y_pred) * 100, 2))
    ROC_AUC.append(round(roc_auc_score(y_test, y_pred) * 100, 2))

    print(clfs[1])
    print ('Confusion Matrix:')
    print(confusion_matrix(y_test, y_pred))
    print("Accuracy : ", accuracy_score(y_test, y_pred) * 100)
    print("Recall : ", recall_score(y_test, y_pred) * 100)
    print("Precision : ", precision_score(y_test, y_pred) * 100)
    print("F1 : ", f1_score(y_test, y_pred) * 100)
    print("ROC-AUC Score - ",roc_auc_score(y_test,y_pred) * 100)
    # print("Classification Report - ",classification_report(y_test,y_pred) * 100)
    print("\n")
```
